```python
import requests
from bs4 import BeautifulSoup
```

```python
def obtenir_contenu_wikipedia(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Erreur {response.status_code} lors de la récupération du contenu.")
        return None
```

```python
def extraire_titre(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    title_tag = soup.find('title')
    if title_tag:
        return title_tag.text
    else:
        return None
```

```python
def extraire_paragraphes(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    result = {}

    for heading in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
        title = heading.text.strip()
        paragraphs = []
        next_element = heading.find_next_sibling()

        while next_element and next_element.name not in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            if next_element.name == 'p':
                paragraphs.append(next_element.text.strip())
            next_element = next_element.find_next_sibling()

        result[title] = paragraphs

    return result
```

```python
def extraire_liens_wikipedia(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    wikipedia_links = []

    for link in soup.find_all('a', href=True):
        if link['href'].startswith('/wiki/'):
            wikipedia_links.append(link['href'])

    return wikipedia_links
```

```python
def analyser_page_wikipedia(url):
    html_content = obtenir_contenu_wikipedia(url)

    if html_content:
        titre = extraire_titre(html_content)
        paragraphes = extraire_paragraphes(html_content)
        liens_wikipedia = extraire_liens_wikipedia(html_content)

        resultat = {
            'titre': titre,
            'paragraphes': paragraphes,
            'liens_wikipedia': liens_wikipedia
        }

        return resultat
    else:
        return None
```

```python
url_test = "https://fr.wikipedia.org/wiki/Intelligence_artificielle"
resultat_test = analyser_page_wikipedia(url_test)

if resultat_test:
    print("Titre de l'article :", resultat_test['titre'])
    print("\nParagraphes et titres associés :")
    for titre, paragraphes in resultat_test['paragraphes'].items():
        print(f"\n{titre}:")
        for paragraphe in paragraphes:
            print(f"- {paragraphe}")
    print("\nLiens vers d'autres pages Wikipédia :", resultat_test['liens_wikipedia'])
else:
    print("Erreur lors de l'analyse de la page.")
```
